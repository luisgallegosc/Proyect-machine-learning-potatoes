### creacion del archivo.csv

```python
def cargar_imagen(ruta):
    imagen = Image.open(ruta)
    # Realizar cualquier preprocesamiento necesario aquí
    # Por ejemplo, cambiar el tamaño, normalizar píxeles, etc.
    imagen = imagen.resize((224, 224))  # Ajusta el tamaño según sea necesario
    imagen = np.array(imagen)
    return imagen
```

```python
carpeta_imagenes = r'C:\Users\benito\OneDrive\Escritorio\UNSAAC\8vo\machine learning\proyecto final\proyect final\CORTES\S17E02'
extensiones_validas = ['.jpg', '.png', '.jpeg']  # Ajusta según las extensiones de tus imágenes

datos = []

for archivo in os.listdir(carpeta_imagenes):
    if archivo.lower().endswith(tuple(extensiones_validas)):
        ruta_completa = os.path.join(carpeta_imagenes, archivo)
        imagen = cargar_imagen(ruta_completa)
        datos.append({'ruta': ruta_completa, 'imagen': imagen})
4
```

```python
df = pd.DataFrame(datos)
df.to_csv('datos_imagenes.csv', index=False)
```
